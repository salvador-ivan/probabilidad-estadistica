##   Histograma

Datos: Temperaturas, en °F, medidas a las 6:00 p.m. durante 35 días.



```r
# Ingresamos los datos:
temperatura<-c(72,78,86,93,106,107,98,82,81,77,87,82,91,95,92,83,76,78,73,81,86,92,93,84,107,99,94,86,81,77,73,76,80,88,91)
```

```r
# formamos el histograma
hist(temperatura)
```

El histograma es diferente al construido en clase debido a que por default el histograma se forma con intervalos de la forma (a, b], para los intervalos que son de la forma [a, b), debemos agregar: right=FALSE

```r
hist(temperatura,right=FALSE)
```

Veamos ahora que ocurre con la instrucción:

```r
hist(temperatura,main="Histograma",xlab="Temperatura",ylab="Frecuencia",col="red",right=FALSE)
```

Con main="Histograma" cambiamos el titulo del gráfico (en este caso a Histograma), xlab="Temperatura" cambia el nombre del eje horizontal (en este caso a Temperatura), ylab="Frecuencia" cambia el nombre del eje horizontal (en este caso a Frecuencia), col="red" agrega color a la gráfica (en este caso rojo).

## cuartiles

```r
# primer cuartil 
quantile(puntos,0.25)
```

```r
# segundo cuartil
quantile(puntos,0.5)
```

```r
# tercer cuartil
quantile(puntos,0.75)
```

```r
# una forma más rápida de obtener los cuartiles es con “fivenum”
fivenum(puntos)
```

con fivenum obtenemos 5 números que siempre serán, en este orden: valor mínimo, primer cuartil, segundo cuartil, tercer cuartil, valor máximo

## Gráfica de caja y bigote (box plot)

Datos: tiempos en segundos de uso de la unidad central de proceso CPU



```r
# Introducimos los datos
cpu<-c(1.17,1.61,1.16,1.38,3.53,1.23,3.76,1.94,0.96,4.75,0.19,2.41,0.71,0.02,1.59,0.15,0.82,0.47,2.16,2.01,0.92,0.75,2.59,3.07,1.40)
```

```r
# la gráfica de caja y bigote
boxplot(cpu)
```

```r
# para que la gráfica aparezca horizontal:
boxplot(cpu, horizontal=TRUE)
```

```r
# agregamos título y nombre al eje horizontal.
boxplot(cpu,main="tiempos en segundos de uso de la CPU", xlab="segundos", horizontal=TRUE)
```

```r
# para obtener los datos de la gráfica, (ver notas del curso)
boxplot(cpu,plot=FALSE)
```
